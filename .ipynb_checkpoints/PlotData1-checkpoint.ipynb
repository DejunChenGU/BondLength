```python
import os
import numpy
file_location = os.path.join('data', 'distance_data_headers.csv')
distance = numpy.genfromtxt(fname=file_location, dtype='unicode', delimiter=',')
print(distance)
```

```python
headers = distance[0]
data = distance[1:]
data = data.astype(numpy.float)
print(data)
```

```python
import matplotlib.pyplot as plt

plt.figure()

```

```python
sample = headers[1]

plt.figure()
plt.xlabel('Simulations')
plt.ylabel('Distance / angstrom')
fig_1 = plt.plot(data[:,0], data[:,1], label=sample)
plt.legend()

plt.savefig(F'{sample}.png')
```

```python

plt.figure()
plt.xlabel('Simulations')
plt.ylabel('Distance / angstrom')
fig_1 = plt.plot(data[:,0], data[:,1], label=headers[1])
fig_1 = plt.plot(data[:,0], data[:,2], label=headers[2])
fig_1 = plt.plot(data[:,0], data[:,3], label=headers[3])
plt.legend()

plt.savefig(F'three_sample.png')
```

```python
plt.figure()
plt.xlabel('Simulations')
plt.ylabel('Distance / angstrom')
fig_1 = plt.plot(data[:,0], data[:,1], label=headers[1])
plt.figure()
fig_2 = plt.plot(data[:,0], data[:,2], label=headers[2])
plt.figure()
fig_3 = plt.plot(data[:,0], data[:,3], label=headers[3])
plt.legend()

plt.savefig(F'three_sample.png')
```

```python
plt.figure()
plt.xlabel('Simulations')
plt.ylabel('Distance / angstrom')

for col in range(1, len(headers)):
    fig = plt.plot(data[:,0], data[:,col], label=headers[col])
    plt.legend()

plt.savefig(F'all_samples.png')
```

```python
plt.figure()
plt.xlabel('Simulations')
plt.ylabel('Distance / angstrom')

for col in range(1, len(headers)):
    fig = plt.plot(data[0::10,0], data[0::10,col], label=headers[col])
    plt.legend()

plt.savefig(F'all_samples_10.png')
```

```python
plt.figure()
plt.xlabel('Simulations')
plt.ylabel('Distance / angstrom')

for col in range(1, len(headers)):
    fig = plt.plot(data[0::100,0], data[0::100,col], label=headers[col])
    plt.legend()

plt.savefig(F'all_samples_100.png')
```

```python
plt.figure()
plt.xlabel('Simulations')
plt.ylabel('Distance / angstrom')

for col in range(1, len(headers)):
    fig = plt.plot(data[0::100,0], data[0::100,col], '--d', label=headers[col])
    plt.legend()

plt.savefig(F'all_samples_bd.png')
```
