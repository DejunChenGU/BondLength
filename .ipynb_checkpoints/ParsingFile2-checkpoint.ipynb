```python
import os

outfile = os.path.join('data', 'sapt.out')
openfile = open(outfile, 'r')
data = openfile.readlines()
openfile.close()
print(len(data))
```

```python
energy_line = []
for line in data:
    if kcal/mol in line:
        energy_line.append(line)
    print(len(energy_line))
```

```python
energy_line = []
for line in data:
    if 'kcal/mol' in line:
        energy_line.append(line)
print(energy_line)
```

```python
energy_line = []
for line in data:
    if 'Electrostatics' or 'Exchange' or 'Induction' or 'Dispersion' in line:
        energy_line.append(line)
print(energy_line)
```

```python
energy_line = []
print(energy_line)
```

```python
energy_line = []
for line in data:
    if 'Electrostatics' in line and 'sSAPT' not in line:
        energy_line.append(line)
    if 'Exchange' in line and 'sSAPT' not in line:
        energy_line.append(line)
    if 'Induction' in line and 'sSAPT' not in line:
        energy_line.append(line)
    if 'Dispersion' in line and 'sSAPT' not in line:
        energy_line.append(line)
print(energy_line)
```

```python
energy_line[0].split()
```

```python
total_energy = 0
energy_list = []
for line in energy_line:
    words = line.split()
    energy = [words[0], words[3], words[4]]
    energy_list.append(energy)             
    total_energy = total_energy + float(words[3])
print(energy_list)
print(total_energy)
```
